```python
def test(x):
    if type(x) != str:
        return x
    else:
        return np.NaN
```

```python
df['Energy Supply']=df['Energy Supply'].apply(lambda x:test(x))
df['Energy Supply per Capita']=df['Energy Supply per Capita'].apply(lambda x:test(x))
```

```python
def application_giga(peta):
    if type(peta) != np.NAN:
        return peta * 1000000
    else:
        return False
```

```python
df['Energy Supply']=df['Energy Supply'].apply(lambda x:application_giga(x))
```
