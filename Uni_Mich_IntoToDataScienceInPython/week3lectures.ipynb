### Pivot tables

```python
df = pd.read_csv('week3/universities_ranking.csv')
```

```python
df
```

```python
def gating(rank):
    if (rank>=1) & (rank<= 100):
        return 'FT'
    elif (rank>=101) & (rank<=200):
        return 'ST'
    elif (rank>=201) & (rank<=300):
        return 'TT'
    else:
        return '4T'
```

```python
df['Rank_level']=df['ranking'].apply(lambda x: gating(x))
```

```python
df
```

```python
df.pivot_table(values='score',index='location',columns='Rank_level',aggfunc=[np.mean]).head()
```
